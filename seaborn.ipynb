# Seaborn for Data Visualization

```python
# setup the dependencies

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
# this allows setting the backend of matplotlib
# to the 'inline' backend: With this backend, the output of plotting
# is displayed inline within froontends (i.e. Jupyter notebook)
import seaborn as sns
print("Setup Complete")
```

```python
# load the data from the csv file into pandas dataframe

fifa_filepath = "./data/fifa.csv"

# read the file and store it in a pandas dataframe object.
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)

### if I don't set the parse_dates=True, pandas doesn't recognize that the index column represents
### real dates, and it doesn't prettify the horizontal axis ticks when plotting
```

```python
fifa_data.head()
```

```python
# set the width/height of the figure
plt.figure(figsize=(16,6))

# draw line chart using seaborn library
sns.lineplot(data=fifa_data)
```

## Line Charts

```python
# using spotify data
spotify_filepath = "./data/spotify.csv"

# read data into dataframe
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
```

```python
spotify_data.head(10)
```

```python
# set width/height of figure
plt.figure(figsize=(14,6))

# add title
plt.title("Daily Global Streams of Popular Songs in 2017-18")

# line chart
sns.lineplot(data=spotify_data)
```

## Plot a subset of the data

```python
list(spotify_data.columns)
```

```python
# plotting just the first 2 songs (shape of you and despacito)
plt.figure(figsize=(14,6))
plt.title("Shape of You vs. Despacito in 2017-18")
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")

# adding label for horizontal axis
plt.xlabel("Date")
```
